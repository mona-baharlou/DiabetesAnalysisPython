```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
num_of_zero_each_col = (data == 0).sum()
num_of_zero_each_col
```

```python
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
data[zero_cols] = data[zero_cols].replace(0 , pd.NA)
data.isnull().sum()
```

```python
data.fillna(data.mean(), inplace = True)
```

```python
data.hist(bins=10, figsize=(12,10))
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
correl = data.corr()
sb.heatmap(correl , annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sb.boxplot(x="Outcome", y="Glucose", data = data)
plt.title('Glucose Comparison Based on Diabetes')
plt.show()
```

```python
sb.boxplot(x="Outcome", y="BMI", data = data)
plt.title('BMI Comparison Based on Diabetes')
plt.show()
```
